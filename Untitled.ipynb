```octave
zoo_data = load('zoo.mat');
```

```octave
[zoo_clusters,cluster_centers] = kmeans(zoo_data.data,7);
```

```octave
zoo_silhouettes = CalcSilhouettes(double(zoo_data.data),zoo_clusters);
```

```octave
[sbc,pci] = PartitionSilhouette(zoo_silhouettes,zoo_clusters);
```

```octave
GraphSilhouetteClusters(zoo_data.labels,zoo_silhouettes,zoo_clusters)
```
